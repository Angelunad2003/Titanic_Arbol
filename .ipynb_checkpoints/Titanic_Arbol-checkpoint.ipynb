```python
#Importamos las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
```

```python
#Cargamos el Dataset y las primeras 5 lineas
data = pd.read_csv("Titanic-Dataset.csv")
```

```python
data.head()
```

```python
#Analizamos las columnas con datos faltantes
data.isnull().sum()
```

```python
#Eliminamos la columna de Cabin
data.drop('Cabin', axis=1, inplace=True)
```

```python
#Eliminamos los datos faltantes de sus respectivas columnas
data.dropna(inplace=True)
```

```python
#Evaluamos los cambios
data.head()
```

```python
le = LabelEncoder()
data['Sex'] = le.fit_transform(data['Sex'])
```

```python
X = data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
y = data['Survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=65)
```

```python
dt = DecisionTreeClassifier(max_depth=3)
dt_model = dt.fit(x_train, y_train)
```

```python
fig = plt.figure(figsize=(12, 10))
tree.plot_tree(dt_model, feature_names=list(X.columns), class_names=['Not survived', 'Survived'])
plt.show()
```

```python
print("Accuracy score on Testing set: ", dt_model.score(x_test, y_test))
```
